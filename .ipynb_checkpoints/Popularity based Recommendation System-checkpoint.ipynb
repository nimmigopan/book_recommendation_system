## Popularity based recommendation

```python
df = merge_df.drop(columns=['ISBN','Year-Of-Publication','Image-URL-S','Image-URL-L','User-ID'])
```

```python
df.head()
```

```python
df.groupby('Book-Title').count()
```

```python
df.groupby('Book-Title').count()['Book-Rating']
```

```python
df.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_of_rating_df = df.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_of_rating_df.rename(columns={'Book-Rating':'no.of.ratings'},inplace=True)
```

```python
num_of_rating_df
```

```python
#finding avg rating of each books
df.groupby('Book-Title').mean()['Book-Rating']
```

```python
mean_of_rating_df = df.groupby('Book-Title').mean().reset_index()
```

```python
mean_of_rating_df.rename(columns={'Book-Rating':'mean.of.ratings'},inplace=True)
```

```python
mean_of_rating_df
```

```python
popularity_df = num_of_rating_df.merge(mean_of_rating_df,on='Book-Title')
```

```python
popularity_df['mean.of.ratings'] = round(popularity_df['mean.of.ratings'],2)
```

```python
popularity_df
```

```python
popularity_df.sort_values('no.of.ratings',ascending=False)
```

```python
popularity_df = popularity_df[popularity_df['no.of.ratings'] > 200]
```

```python
popularity_df
```

```python
popularity_df = popularity_df.merge(df,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df.drop(columns=['Book-Rating'],inplace=True)
```

```python
popularity_df.sort_values('mean.of.ratings',ascending=False).head(100)
```

# Collobrative based recomm.system

```python
merge_df
```

```python
df = merge_df.drop(columns=['ISBN','Year-Of-Publication','Image-URL-S','Image-URL-L'])
```

```python
df.head()
```

```python
df['User-ID'].value_counts() #92106 unique users
```

```python
df.groupby('User-ID').count()
```

```python
df.groupby('User-ID').count()['Book-Rating']
```

```python
df.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
 x= df.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
x[x] #1648 users out of 92106 gave rating for more than 100 books
```

```python
x[x].index
```

```python
temp_df = x[x].index
```

```python
df['User-ID'].isin(temp_df)
```

```python
df[df['User-ID'].isin(temp_df)]
```

```python
filtered_on_bases_of_users_df = df[df['User-ID'].isin(temp_df)]
```

```python
# 1031129 rows in 'merge_df' filtered to 591891 rows on the bases of users who rated for more than 100 books
```

```python
filtered_on_bases_of_users_df
```

```python
# filter on bases of popular books
filtered_on_bases_of_users_df.groupby('Book-Title').count()
```

```python
filtered_on_bases_of_users_df.groupby('Book-Title').count()['Book-Rating']
```

```python
filtered_on_bases_of_users_df.groupby('Book-Title').count()['Book-Rating'] > 50
```

```python
y = filtered_on_bases_of_users_df.groupby('Book-Title').count()['Book-Rating'] > 50
```

```python
y[y]
```

```python
y[y].index
```

```python
temp_df = y[y].index
```

```python
final_df = filtered_on_bases_of_users_df[filtered_on_bases_of_users_df['Book-Title'].isin(temp_df)]
```

```python
final_df
```

```python
pt = final_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt #679 books & 810 users
```

```python
pt.fillna(0,inplace=True)
```
