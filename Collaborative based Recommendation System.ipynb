```python
final_df = filtered_on_bases_of_users_df[filtered_on_bases_of_users_df['Book-Title'].isin(temp_df)]
```

```python
final_df
```

```python
pt = final_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt #679 books & 810 users
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# each books represented as a vector in 810 dimensional space.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt) # A 2D numpy array that gives Euclidean distance of each book with 679 other books
```

```python
cosine_similarity(pt).shape
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    #fetchind index
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```

```python
recommend('Zoya')
```

```python
recommend("Harry Potter and the Sorcerer's Stone (Book 1)")
```
