## Popularity based recommendation

```python
df = merge_df.drop(columns=['ISBN','Year-Of-Publication','Image-URL-S','Image-URL-L','User-ID'])
```

```python
df.head()
```

```python
df.groupby('Book-Title').count()
```

```python
df.groupby('Book-Title').count()['Book-Rating']
```

```python
df.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_of_rating_df = df.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_of_rating_df.rename(columns={'Book-Rating':'no.of.ratings'},inplace=True)
```

```python
num_of_rating_df
```

```python
#finding avg rating of each books
df.groupby('Book-Title').mean('Book-Rating')
```

```python
mean_of_rating_df = df.groupby('Book-Title').mean('Book-Rating').reset_index()
```

```python
mean_of_rating_df.rename(columns={'Book-Rating':'mean.of.ratings'},inplace=True)
```

```python
mean_of_rating_df
```

```python
popularity_df = num_of_rating_df.merge(mean_of_rating_df,on='Book-Title')
```

```python
popularity_df['mean.of.ratings'] = round(popularity_df['mean.of.ratings'],2)
```

```python
popularity_df
```

```python
popularity_df.sort_values('no.of.ratings',ascending=False)
```

```python
popularity_df = popularity_df[popularity_df['no.of.ratings'] > 200]
```

```python
popularity_df
```

```python
popularity_df = popularity_df.merge(df,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df.drop(columns=['Book-Rating'],inplace=True)
```

```python
popularity_df.sort_values('mean.of.ratings',ascending=False).head(100)
```
